```python
import numpy as np

def compresschar(input_str):
    result = "" # Initialize the result string
    current_char = "" # Initialize the current character and count
    count = 0

    for char in input_str:
        if char == current_char: #If the character is the same as the current one, increment the count
            count += 1

        else:

            if current_char:   # If the character is different, add the current character and count to the result
                if count>1:
                  result += current_char + str(count)
                else:
                  result = result + current_char

            current_char = char   # Update the current character and reset the count
            count = 1


    if current_char: # Add the last character and its count to the result
        if count>1:
          result += current_char + str(count)
        else:
          result = result+current_char

    return result



input_string =input("Enter String : ")
output_string = compresschar(input_string)
print(output_string)
```
